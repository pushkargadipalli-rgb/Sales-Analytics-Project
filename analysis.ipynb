```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")
```

```python
df = pd.read_csv(r"C:\sales_data.csv", encoding="latin1")
```

```python
df.head()
```

```python
df.shape
df.info()
df.describe()
```

```python
plt.figure(figsize=(8,5))

sns.barplot(
    x="Category",
    y="Sales",
    data=df,
    estimator=sum,
    errorbar=None,
    hue="Category",
    palette="Blues_d",
    legend=False
)

plt.title("Total Sales by Category")
plt.ylabel("Total Sales")
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.barplot(
    x="Category",
    y="Profit",
    data=df,
    estimator=sum,
    errorbar=None,
    hue="Category",
    palette="Greens_d",
    legend=False
)

plt.title("Total Profit by Category")
plt.ylabel("Total Profit")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.barplot(
    x="Region",
    y="Sales",
    data=df,
    estimator=sum,
    errorbar=None,
    hue="Region",
    palette="Oranges_d",
    legend=False
)

plt.title("Total Sales by Region")
plt.ylabel("Total Sales")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.barplot(
    x="Segment",
    y="Profit",
    data=df,
    estimator=sum,
    errorbar=None,
    hue="Segment",
    palette="Purples_d",
    legend=False
)

plt.title("Total Profit by Segment")
plt.ylabel("Total Profit")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.barplot(
    x="Segment",
    y="Sales",
    data=df,
    estimator=sum,
    errorbar=None,
    hue="Segment",
    palette="coolwarm",
    legend=False
)

plt.title("Total Sales by Segment")
plt.ylabel("Total Sales")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    x="Discount",
    y="Profit",
    data=df
)

plt.title("Discount vs Profit")
plt.xlabel("Discount")
plt.ylabel("Profit")
plt.tight_layout()
plt.show()
```

```python
top_subcat = df.groupby("Sub-Category")["Sales"].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))

sns.barplot(
    x=top_subcat.values,
    y=top_subcat.index,
    hue=top_subcat.index,
    palette="viridis",
    legend=False
)

plt.title("Top 10 Sub-Categories by Sales")
plt.xlabel("Total Sales")
plt.ylabel("Sub-Category")
plt.tight_layout()
plt.show()
```

Conclusion:

The analysis shows that the Consumer segment is the primary driver of both sales and profit. Technology products contribute the highest revenue, especially in sub-categories like Phones and Chairs. However, certain sub-categories such as Tables generate losses due to high discounting. The West region performs best in terms of profit. The business should reduce excessive discounts and focus on high-performing categories to improve profitability.
